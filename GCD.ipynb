## Lemma 2.5.3 proof

$$
gcd(a, n) = c\\
gcd(b, n) = d\\

a - b = nk
$$

---

c is a factor of both a and n   
$$
a - nk = b\\
\frac{a}{c} - \frac{nk}{c} = \frac{b}{c}\\
int - int*k = int = \frac{b}{c}\\
$$

thus c is also a factor of b  
d is the greatest common factor of b and n  

$$
c <= d
$$

---

d is a factor of both b and n

$$
a = nk + b\\
\frac{a}{d} = \frac{nk}{d} + \frac{b}{d}\\
\frac{a}{d} = k*int + int = int\\
$$

thus d is also a factor of a  
c is the greatest common factor of a and n  

$$
d <= c
$$

---

$$
c <= d \\
d <= c \\
c == d
$$

---

Thus if $a \equiv b$ (mod n)  
Then gcd(a, n) = gcd(b, n)

## Euclidean Algorithm

$$
gcd(n, a) = gcd(a, r)\\

gcd(a, r) = gcd(r, r')\\

gcd(a, 0) = |a|
$$

````python
def gcd(n, r):
    if (r == 0):
        return n
    return gcd(r, n % r)

gcd(180, 160), gcd(181, 18)
````

## Modular Inversion Theorem 2.5.14

There exists $x$ s.t. $ax \equiv 1$ (mod n)

AKA $x = a^{-1}$ (mod n)

IFF 

$gcd(a, n) = 1$

Moreover

For Bezoat Coeff (x, y)

x is AN inverse fo a mod n

### Proof

$$
ax === 1 \text{   (mod n)}
$$

(ax - 1) is a multiple of n

$$
ax - 1 = ny\\
ax - ny = 1
$$

gcd(a, n) is a factor of both a and n [by defn of gcd]

so (ax - ny) must be divisible by gcd(a, n) as well

so (1) must also be divisible by gcd(a, n)

But the only divisor of 1 is 1

Thus gcd(a, n) = 1

---

Conversely

$$
gcd(a, n) = 1 
$$

implies existence of Bezout coeffs

$$
ax + ny = 1\\
ax - 1 = ny
$$

(ax - 1) is a multiple of n 

$$
ax === 1 \text{   (mod n)}
$$

Thus with both sides proven,

$$ x \equiv a^{-1} \text{   (mod n)}$$

## Lemma 2.5.17

Modular inverses are "unique" under mod n

If $x$ and $x'$ are both inverses of $a$ mod n,

Then $x \equiv x'$ (mod n)

$$
\begin{align*}
x &=   (1)x\\
  &\equiv (x'a)x\\
  &\equiv x'(ax)\\
  &\equiv x'(a)\\
  &= x' \qquad\qquad \text{(mod n)}
\end{align*}
